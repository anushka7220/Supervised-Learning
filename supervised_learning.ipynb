## This is the linear regression model showing sales of cookies with respect to the temperature
### Import necessary libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

Lets prepare a dataset

```python
#create a dataset
data= {
    'Temperature (°C)': [20, 22, 25, 27, 30, 32, 35, 37, 40],
    'Cookies Sold': [500, 450, 400, 350, 300, 250, 200, 180, 150]
}
#converting it to a dataframe
df = pd.DataFrame(data)
#lets display it
print(df)
```

```python
# Scatter plot of the data
plt.scatter(df['Temperature (°C)'], df['Cookies Sold'], color='brown')
plt.title('Temperature vs Cookies Sold')
plt.xlabel('Temperature (°C)')
plt.ylabel('Cookies Sold')
plt.gca().set_facecolor('wheat')
plt.grid(True)
plt.show()
```

```python
# Separate features (X) and target (Y)
X = df[['Temperature (°C)']].values
Y = df['Cookies Sold'].values

# Split into training and testing sets (80% train, 20% test)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

print("Training data points:", len(X_train))
print("Testing data points:", len(X_test))
```

```python
# Create and train the model
model = LinearRegression()
model.fit(X_train, Y_train)

# Get the model parameters
print("Slope (m):", model.coef_[0])
print("Intercept (b):", model.intercept_)
```

```python
# Predict on test set
Y_pred = model.predict(X_test)

# Compare actual vs predicted values
results = pd.DataFrame({'Actual': Y_test, 'Predicted': Y_pred})
print(results)
```

```python
# Calculate Mean Squared Error
mse = mean_squared_error(Y_test, Y_pred)
print("Mean Squared Error (MSE):", mse)
```

```python
# Scatter plot of the original data
plt.scatter(X, Y, color='brown', label='Actual data')

# Plot the regression line
plt.plot(X, model.predict(X), color='black', label='Regression line')

plt.title('Linear Regression: Temperature vs Cookies Sold')
plt.xlabel('Temperature (°C)')
plt.ylabel('Cookies Sold')
plt.gca().set_facecolor('wheat')
plt.legend()
plt.grid(True)
plt.show()
```
